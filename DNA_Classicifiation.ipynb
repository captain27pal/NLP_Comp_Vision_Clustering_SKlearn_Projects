```python
import pandas as pd
import numpy as np
import sys
print(f'pandas version: {pd.__version__}')
print(f'numpy version: {np.__version__}')
print(f'sys version: {sys.version}')
```

```python
url='https://archive.ics.uci.edu//ml//machine-learning-databases//molecular-biology//promoter-gene-sequences//promoters.data'

names=['class','id','sequence']
data=pd.read_csv(url,names=names)
```

```python
data.info()
```

```python
data.head()
```

```python
data.iloc[0:4]
#the backslash \t is due to the space in the data set and therefore we need to remove them from the dataset
```

```python
#so, lets start the data cleaning process
classes=data.loc[:,'class']
classes[0:4]

dataset={}
sequence=list(data.loc[:,'sequence'])
print(sequence[0:5])
type(sequence)
```

```python
for i,seq in enumerate(sequence):
    neucleotides=list(seq)
    neucleotides=[x for x in neucleotides if x!='\t']
    
    neucleotides.append(classes[i])
    dataset[i]=neucleotides
print(dataset[19])
```

```python
data=pd.DataFrame(dataset)
data.info()
data.head()
#Since the new dataframe has the shape of 57*105 we need to transpose it to get the original shape of 105*57. to do that use the
#transpose method of pandas..simple
```

```python
df=data.transpose()
df.info()
#now the df dataframe object has 58 columns(0 to 57) and 106 non  null rows.. we are now good to go
```

```python
df.iloc[100:105]
# there is still a problem. the columns are named as integer. It does not really matter for the rest, but for the class column we would like to get it
# renamed as 'class'
```

```python
df.rename(columns={57:'class'},inplace=True)# inplace=True means do not generate a copy, make the changes in the original dataframe
```

```python
# now see, the last column has been name as 'class'
df.describe()
```

```python
#record value counts for each sequence
series=[]
for name in df.columns:
    series.append(df[name].value_counts())
info=pd.DataFrame(series)
print(info.transpose())
```

```python
#now lets convert the data into numerical data using pd.get_dummies() method which will convert th categorical variables into numeric ones by doing onehot encoding
# in this case since we have 4 types of proties, the number of columns will be (57*4)=228 + 1(for +)+1(for -) 230 columns for each row
# this is one of the important pandas methods
df=pd.get_dummies(df)
df
```

```python
# one little cleaning. we can drop the class- column as even if e have only class + column it will have 0 for the rows where its not 1
# so, we will get 1 for positive and 0 for negative even if we keep only the class+ column
df.drop(['class_-'],axis=1,inplace=True)
df
```

```python
# now we can rename the class_+ column as class
df.rename(columns={'class_+':'class'},inplace=True)
df
```

```python
df.sample(frac=1).reset_index(drop=True)
df.iloc[59]
```

```python
#we can also use sklearn shuffle
from sklearn.utils import shuffle
```

```python
df=shuffle(df).reset_index(drop=True)
df
# we shuffled the data as the + and - casses are devided in the first and second half of the data, which might get the model
# learn that posibility of + is higher if the case is in the firsthalf of the dataset. So, we built some randomness by shuffling'
# the data by using the sklearn shuffle function. .rest_index(drop=True) argument ensures that the original indexes are droppedd and
# we get a serial indexing
```

```python
# so now we have only numerial data in our data and is shuffled. So, we can now move on with fitting the data into our models
#lets import the models
from sklearn.neighbors import KNeighborsClassifier
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import classification_report,accuracy_score
```

```python
#split the data into train and test
#first drop the class column from the train data, 
from sklearn.model_selection import train_test_split
seed=1
X=np.array(df.drop(['class'],axis=1))
Y=np.array(df['class'])
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=seed)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
#define scoring
scoring='accuracy'
# deifne the models

names=['KNeighbors','MLPCclassifier','Gaussianprocessclassifier','Decisiontree','Randomforest','Adaboost','GaussianNB','SVC-linear','SVC-rbf','SVC-sigmoid']
classifiers=[
    KNeighborsClassifier(n_neighbors=3),
    MLPClassifier(),
    GaussianProcessClassifier(1.0*RBF(1.0)),
    DecisionTreeClassifier(max_depth=3),
    RandomForestClassifier(max_depth=3,n_estimators=10,max_features = 1),
    AdaBoostClassifier(),
    GaussianNB(),
    SVC(kernel='linear'),
    SVC(kernel='rbf'),
    SVC(kernel='sigmoid')
    
]
models=zip(names,classifiers)
```

```python
#now import kfold for doing something similar to what cross_val function would have done where in we define cv=no of splits
# and the  it spilits the dataset into that many folds(eg 10), train in 9 folds and test in the 10th fold, and repeat the process 10 times
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
kfold=KFold(n_splits=10,random_state=seed,shuffle=True)
```

```python
#evaluate each model
results=[]
names=[]
for name, model in models:
    kfold=kfold
    cv_scores=cross_val_score(model,X_train,Y_train,cv=kfold,scoring=scoring,)
    results.append(cv_scores)
    names.append(name)
    model.fit(X_train,Y_train)
    predictions=model.predict(X_test)
    print(f'{name}: {cv_scores.mean()} with std: {cv_scores.std()}')
    print('==========================')
    print('Performance on test data:')
    print(f'accuracy:{accuracy_score(Y_test,predictions)}')
    print(classification_report(Y_test,predictions))
    print('*******end********')
```
